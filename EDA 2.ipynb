##  Replacing null values 
## 1) imputation i.e we fill the null values with mean , median and mode
## 2) predictive modeling i.e the model predicts the null value 

```python
df_health[cat].isnull().sum()
```

```python
df_health[num].isnull().sum()
```

```python
for i in cat:
    mode = df_health[i].mode()[0]
    df_health.fillna(mode,inplace = True)
```

```python
df_health[cat].isnull().sum()
```

```python
for i in num:
    mean = df_health[i].mean()
    df_health.fillna(mean,inplace=True)
```

## Dendrogram

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
wardlink = linkage(df_health,method='ward')
```

```python
plt.figure(figsize=(21,10))
dend = dendrogram(wardlink)
```

```python
dend1 = dendrogram(wardlink,
                   truncate_mode='lastp',p= 10) ## trucate means it took the last ten points nd prepared a dendogram
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
clusters = fcluster(wardlink,2,criterion='maxclust')
```

```python
clusters
```

```python
df_health['hier_clusters'] = clusters
```

## Elbow Chart

```python
Inertias = []
```

```python
for i in range (1,11):
    KM = KMeans(n_clusters=i)
    KM.fit(df_health_sca)
    Inertias.append(KM.inertia_)
    
```

```python
Inertias
```

```python
## The above code can be done in this way also where here i have not mentioned variable to kmeans 

for i in range(1,11):
    KMeans(n_clusters=i).fit(df_health_sca)
    Inertias.append(KM.inertia_)
```
